```python
# Import libraries.
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
sns.set(style = 'white')
```

```python
# the function has actually been used a lot in this book, but after calling it and studying
# the output, the cell is deleted

#makes displaying large dateframes manageable
def display_all(df):
    '''temporarily increase display max and transpose the dateframe view
    for enhanced readability
    '''
    with pd.option_context("display.max_rows", 100, "display.max_columns", 100): 
        display(df.head().transpose())
```

```python
# displays two key categorical column charts for quick analysis
def cat_info(col):
    """function takes a dataframe and its categorical column, and displays two
    horizontal bar chart, aligned by category: value counts chart and mean price 
    per category chart
    """
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14,4))
    fig.suptitle(col, size=20, color='maroon', y=1)
    # value counts on the left chart
    ames[col].value_counts(sort=False).sort_index().plot(kind='barh', ax=ax1, color='darkorchid')
    ax1.set_title('Value counts', fontsize=14, color='darkgoldenrod')
    
    
    # mean sale price per category
    ames.sort_values(col).groupby(col)['saleprice'].mean().plot(kind = 'barh', ax=ax2
                                                                , color='olivedrab')
    ax2.set_title('Avg Sale Price', fontsize=14, color='darkgoldenrod')
    ax2.set_ylabel(None)
    ax2.set_yticks([])
    ax2.axvline(x=ames['saleprice'].mean(), dash_capstyle='butt', linestyle='-.')
```

```python
# ames is train, ames2 is test dataset
ames = pd.read_csv('../datasets/train.csv')
ames2 = pd.read_csv('../datasets/test.csv')
```

* [Back to top](#Content)

```python
display_all(ames)
```

```python
#drop columns with fewer than 10% of non-nulls
for col in ames.columns:
    if ((ames[col].isnull().sum())/len(ames[col])) > 0.9:
        ames.drop(col, axis=1, inplace=True)
        
for col in ames2.columns:
    if ((ames2[col].isnull().sum())/len(ames2[col])) > 0.9:
        ames2.drop(col, axis=1, inplace=True)
```

```python
ames.shape
#3 columns dropped
```

```python
#simplify column names
ames.columns = [x.lower().replace(' ', '') for x in ames.columns]

ames2.columns = [x.lower().replace(' ', '') for x in ames2.columns]
```

```python
num_col, cat_col = [], []
for col in ames2.columns:
    if pd.api.types.is_numeric_dtype(ames[col]):
        num_col.append(col)
    else:
        cat_col.append(col)

 
```

```python
# check the categorical column nulls
ames2[cat_col].isnull().sum()
```

#### After studying the data dictionary, it became apparent that all the null values should actually be "None"s, ie it's not a case of missing value, but rather it's a category "None". That holds for all the non-nulls in the categorical columns above.

```python
# replace nan with "None"
for col in cat_col:
    ames[col].fillna("None", inplace=True)
    ames2[col].fillna("None", inplace=True)
```

### 3.0 In-depth analysis of categorical features
* [Back to top](#Content)

### To speed up data cleaning and exploration, the decision has been made to focus on categorical columns first.  The flow will be to use the function cat_info created above to chart value_counts and price per category for all the categorical columns, 7 at a time, and then go down the list and decide what to do with them. There are two concurrent , yet sometimes opposing, objectives at work here: 
- keep all the relevant data that can help us predict sale value
- keep the number of features as low as possible to enable a crisp, concise and interpretable machine learning

### 3.1 Batch 1
* [Back to top](#Content)

```python
for col in cat_col[0:7]:
    cat_info(col)
```

### **Conclusions (1-7):**
- **mszoning:** important, it affects the price, has 3 substantial categories (RM, RL, FV). Dummify all 3, the rest will be 'baseline'
- **street:** virtually a one-value category. easy drop
- **lotshape:** combine 3 irregular values (ir1, ir2 are ir3) into 1 dummy, as they command a higher sale price than reg. (All 3 appear to be just various degrees of odd shapes, according to the data dictionary, thus combinable)
- **landcontour:** some price effect noticeable for hillside contour, but small value counts make the category ineffective from feature clutter vs price discovery standpoint. Drop the column.
- **utilites:** virtually a one-value column. Drop.
- **lotconfig:** only CulDSac has predictive power, with high mean price. Small count  for most values and most values don't effect the price much however. We have to cut aggresively as we have too many features. It's a tough call, but drop.
- **landslope:** very one-sided feature. drop.

```python
# get dummies for the zoning column, only for the values mentioned above,
# the rest will be 'baseline'
ames = pd.get_dummies(ames, columns=['mszoning'], prefix="zone")
ames.drop(['zone_RH', 'zone_I (all)', 'zone_C (all)', 'zone_A (agr)'], axis=1, inplace=True)

ames2 = pd.get_dummies(ames2, columns=['mszoning'], prefix="zone")
ames2.drop(['zone_RH', 'zone_I (all)', 'zone_C (all)'], axis=1, inplace=True)
```

```python
# create one dummy for all irregular lot shapes
ames['lotshape_irr'] = np.where((ames['lotshape']=='IR1')
                                | (ames['lotshape']=='IR2') | (ames['lotshape']=='IR3'), 1, 0)

ames2['lotshape_irr'] = np.where((ames2['lotshape']=='IR1')
                                | (ames2['lotshape']=='IR2') | (ames2['lotshape']=='IR3'), 1, 0)
```

```python
ames.drop('lotshape', axis=1, inplace=True)
ames2.drop('lotshape', axis=1, inplace=True)
```

### **Conclusions (8-14):**
- **neighborhood:** busy but important column. study how to keep the column without exploding the number of features with 27 new dummies.
- **condition1:** proximity to arterial or feeder negatively affects the price. makes sense. create one dummy out of those two values. change the column name
- **condition2:** virtually a one-value column. Drop.
- **bldgtype:** townhouse, duplex, 2 fam conversion are priced lower. Turn those 3 into one dummy.
- **housestyle:** 2+ stories command a higher price. Makes sense. Turn them into 1 dummy
- **roofstyle:** hip roof commands a premium. dummify it.
- **roofmatl:** very one-sided feature. drop.

#### Rank all 28 neighboorhoods based on their desirability, manually, using https://bestneighborhood.org/best-neighborhoods-ames-ia/, google.maps, and other real-estate websites to locate the 'neighborhood' column values on the map.
- using desirability index rather than price index to rank the neighborhoods is important. If we were using the price, then we'd be predicting the sales price based on the .... sales price.
- the problem, however, might be that a 'desirable' neighborhood is nothing more than a 'high priced' one, in which case "using the price to predict the price" problem is still there. We must, however, take stock of the fact that the real-estate is a "reflexive" market, to use the term coined by the famous investor George Soros. In other words, "desirable" neighborhoods are high-priced, and the high-priced neighborhoods are desireable, creating a positive feedback loop.This essentially guarantees that price is the predictor of price when it comes to real estate valuation. It is however beyond the scope of this notebook to address, adjust or correct this phenomenon.

```python
# aftter(many hours of) charting the Ames house values, the result was this ordered array,
#  where every neighborhood is paired up with 1-3 value (3 being most desireable).
codes = sorted(ames['neighborhood'].unique())
ranks = [2,2,1,1,3,1,2,3,2,3,3,2,2,1,1,1,2,2,3,3,1,1,1,3,3,3,3,3]
hood_dict = dict(zip(codes, ranks))
```

```python
# assinged every neighborhood a 1-3 value based on its desireabilty. (3=highest)
ames['neighborhood'] = ames['neighborhood'].map(hood_dict)
ames2['neighborhood'] = ames2['neighborhood'].map(hood_dict)
```

```python
# lump feedr and artery into one dummy
ames['near_busy_st'] = np.where((ames['condition1']=='Feedr') | (ames['condition1']=='Artery'),
                                1, 0)
ames2['near_busy_st'] = np.where((ames2['condition1']=='Feedr') | (ames2['condition1']=='Artery'),
                                1, 0)
```

```python
# lump non-single family homes into one dummy
ames['non_1fam'] = np.where((ames['bldgtype']=='Twnhs') | (ames['bldgtype']=='Duplex')
                            | (ames['bldgtype']=='2fmCon'), 1, 0)

ames2['non_1fam'] = np.where((ames2['bldgtype']=='Twnhs') | (ames2['bldgtype']=='Duplex')
                            | (ames2['bldgtype']=='2fmCon'), 1, 0)
```

```python
# lump taller homes into one dummy
ames['2plus_fl'] = np.where((ames['housestyle']=='2Story') | (ames['housestyle']=='2.5Unf')
                            | (ames['housestyle']=='2.5Fin'), 1, 0)
ames2['2plus_fl'] = np.where((ames2['housestyle']=='2Story') | (ames2['housestyle']=='2.5Unf')
                            | (ames2['housestyle']=='2.5Fin'), 1, 0)
```

```python
# dummy for 'Hip' style roof
ames['hip_roof'] = np.where((ames['roofstyle']=='Hip'), 1, 0)
ames2['hip_roof'] = np.where((ames2['roofstyle']=='Hip'), 1, 0)
```

```python
ames.drop(['roofstyle', 'housestyle', 'bldgtype', 'condition1'], axis=1, inplace=True)
ames2.drop(['roofstyle', 'housestyle', 'bldgtype', 'condition1'], axis=1, inplace=True)
```

### **Conclusions (15-20):**
- **exterior1st:** a difficult column. "CemntBd" and "VynilSd" are premium exteriors, while "Wd Sdng", "Plywood", "MetalSd" and "HdBoard" are inferior, based on sale price. Dummify those two  groups, use the rest as baseline. Maybe drop the entire column if multi-colinearity surfaces.
- **exterior2nd:** almost a carbon-copy of exterior1st. Drop
- **masvnrtype:**  Turn 3 non-None values into a dummy, as veneer increases the value of the home. ('common brick' actually doesn't, but it's has a very tiny value-count, so it'll barely affect the stats. it's also better than dropping rows just because they don't suit our preconceived notions)
- **exterqual:** good category, but we have too many of these qual/cond columns, all correlated. we have too many features as it is, so drop
- **extercond:** see above. drop.
- **foundation:** 1 dummy - 'PConc', as a premium value, the rest 'baseline'

```python
# turn exterior into an ordered 0,1,2 column, from worst to best
ames['ext_ord'] = ames['exterior1st'].apply(lambda v: 2 if v in ['CemntBd', 'VinylSd'] else
                                           (0 if v in ['Wd Sdng', 'Plywood', 'MetalSd', 'HdBoard']
                                            else 1))
                                            
ames2['ext_ord'] = ames2['exterior1st'].apply(lambda v: 2 if v in ['CemntBd', 'VinylSd'] else
                                           (0 if v in ['Wd Sdng', 'Plywood', 'MetalSd', 'HdBoard']
                                            else 1))
```

```python
# add dummy for any veneer
ames['has_veneer'] = np.where((ames['masvnrtype']=='Stone') | (ames['masvnrtype']=='BrkFace')
                            | (ames['masvnrtype']=='BrkCmn'), 1, 0)

ames2['has_veneer'] = np.where((ames2['masvnrtype']=='Stone') | (ames2['masvnrtype']=='BrkFace')
                            | (ames2['masvnrtype']=='BrkCmn'), 1, 0)
```

```python
# add dummy when foundation is made out of poured concrete
ames['cncrt_found'] = np.where((ames['foundation']=='PConc'), 1, 0)
ames2['cncrt_found'] = np.where((ames2['foundation']=='PConc'), 1, 0)
```

```python
ames.drop(['foundation', 'masvnrtype', 'exterior1st'], axis=1, inplace=True)
ames2.drop(['foundation', 'masvnrtype', 'exterior1st'], axis=1, inplace=True)
```

### **Conclusions (21-28):**
- **bsmtqual:** make one ordinal column out of 4 basement categorical columns by translating value to 1-5 scale, and averaging the row totals.
- **bsmtexposure:** see 'bsmtqual. this category, confusingly, has 'None'(no basement) and 'No' (yes basement, but no exposure. Merge them, since no basement also means no basement exposure.
- **bsmtfintype1:** see 'bsmtqual'
- **bsmtfintype2:** too similar to type1. drop.
- **heating:**  drop. virtually a one-value column
- **heatingqc:** 1 dummy - 'Ex', as a premium value, the rest 'baseline'
- **centralair:** good category, with two distinctly-priced, sizebale values. dummify.

```python
# translate basement codes into ordered numerical values
codes = sorted(ames['bsmtqual'].unique())
ranks = [5,2,4,0,1,3]
hood_dict = dict(zip(codes, ranks))
hood_dict
```

```python
# change the columns 
ames['bsmtqual'] = ames['bsmtqual'].map(hood_dict)
ames['bsmtcond'] = ames['bsmtcond'].map(hood_dict)

ames2['bsmtqual'] = ames2['bsmtqual'].map(hood_dict)
ames2['bsmtcond'] = ames2['bsmtcond'].map(hood_dict)
```

```python
# do the same for basement exposure
codes = sorted(ames['bsmtexposure'].unique())
ranks = [3,5,1,0,0]
hood_dict = dict(zip(codes, ranks))
hood_dict
```

```python
ames['bsmtexposure'] = ames['bsmtexposure'].map(hood_dict)

ames2['bsmtexposure'] = ames2['bsmtexposure'].map(hood_dict)
```

```python
# and finally for the basement type
codes = sorted(ames['bsmtfintype1'].unique())
ranks = [4,3,5,1,0,2,1]
hood_dict = dict(zip(codes, ranks))
hood_dict
```

```python
ames['bsmtfintype1'] = ames['bsmtfintype1'].map(hood_dict)

ames2['bsmtfintype1'] = ames2['bsmtfintype1'].map(hood_dict)
```

```python
# create the bsmt_ord column that combines the data from all the basement columns
ames['bsmt_ord'] = round(ames[['bsmtqual', 'bsmtcond', 'bsmtexposure', 'bsmtfintype1']]
                         .mean(axis=1), 2)

ames2['bsmt_ord'] = round(ames2[['bsmtqual', 'bsmtcond', 'bsmtexposure', 'bsmtfintype1']]
                         .mean(axis=1), 2)
```

```python
ames.drop(columns=['bsmtqual', 'bsmtcond', 'bsmtexposure', 'bsmtfintype1'], inplace=True)
ames2.drop(columns=['bsmtqual', 'bsmtcond', 'bsmtexposure', 'bsmtfintype1'], inplace=True)
```

```python
# central a/c dummy
ames['cent_air_dum'] = np.where((ames['centralair']=='Y'), 1, 0)
ames2['cent_air_dum'] = np.where((ames2['centralair']=='Y'), 1, 0)
```

```python
ames.drop(columns=['heatingqc', 'centralair'], axis=1, inplace=True)
ames2.drop(columns=['heatingqc', 'centralair'], axis=1, inplace=True)
```

### **Conclusions (29-39):**
- **electrical:** mostly a one-value category, likely correlated with others. drop.
- **kitchenqual:** decent size in every value bucket, good price differentiation. encode. (compromise - dummifying would be better, but would create too many columns.)
- **functional:**  drop. virtually a one-value column
- **fireplacequ:** good category with many observations for each value. since both the number and the quality of fireplaces seem highly predictve of the saleprice, yet in order to save column space, we ought to add 'fireplacequ' to 'fireplaces' column (the number of fireplaces) and average them, to reduce feature numbers. We must be very careful how we do it since we are adding a qual and quant column. Make sure that 1 point on the ordinal scale is worth roughly the same dollar increase in the saleprice.
- **garagetype:**  (similar to firplaces - size and quality both matter) encode all garage categorical columns on 1-5 scale, add and average them. add a numerical categorical columns 'garage cars' to the average, as it's a proxy for garage size, and it's already perfectly scaled, from 0-5. Don't include garagecond and garagequal columns as they don't have much variety and thus don't do much to predict the sales price
- **garagefinish:** see 'garagetype'
- **garagequal:** see 'garagetype'. drop.
- **garagecond:** see 'garagetype'. drop.
- **paveddrive:** dummify 'yes', but check the correlation with the other 'paved' column
- **fence:**  not much price variability within the category and not that many houses per value. upon research, fences is an ambigous category. On one hand a fence is nice to have and valuable, but on the other in the USA nice houses in low-crime neighborhoods sometimes do not have one. drop.
- **saletype:** most values in the category have few counts and/or not much effect on price. only 'new' value seems to have predicting power, but not much and can be easily exploited through other categories such as 'age' or 'cond'. Drop


```python
# translate kitchen codes into ordered numerical values
# this cell of code is a bit different than other since, ames2, the smaller test dataset,
# contains the kitchen quality value 'poor', which is not found in ames, the bigger train set
# use ames2, rather than ames to get codes
codes = sorted(ames2['kitchenqual'].unique())
ranks = [3,0,2,0,1]
hood_dict = dict(zip(codes, ranks))
hood_dict
```

```python
ames['kitchenqual'] = ames['kitchenqual'].map(hood_dict)
ames2['kitchenqual'] = ames2['kitchenqual'].map(hood_dict)
```

```python
# learn a bit about the garage numeric column that'll be averaged with garage categorical columns
ames['garagecars'].value_counts(dropna=False)
```

```python
# drop nan row. garages with size for 0 cars are already represented, so nan doesn't make sense.
ames.drop(ames.loc[ames['garagecars'].isnull()].index, inplace=True)
```

```python
# change garagetype codes into ordered numeric values
codes = sorted(ames['garagetype'].unique())
ranks = [0,1,0,2,0,0,0]
hood_dict = dict(zip(codes, ranks))
hood_dict
```

```python
# swap the codes for numeric values
ames['garagetype'] = ames['garagetype'].map(hood_dict)
ames2['garagetype'] = ames2['garagetype'].map(hood_dict)
```

```python
# do the same for garage finish
codes = sorted(ames['garagefinish'].unique())
ranks = [2,0,2,1]
hood_dict = dict(zip(codes, ranks))
hood_dict
```

```python
ames['garagefinish'] = ames['garagefinish'].map(hood_dict)
ames2['garagefinish'] = ames2['garagefinish'].map(hood_dict)
```

```python
#check if 'garagecars' columns needs transforming
cat_info('garagecars')
```

```python
# combine all the huge garages with space for 3 more cars into 1 category '3'
# as the ones with room for more than 3 cars don't add extra value to the house
ames['garagecars'] = ames['garagecars'].apply(lambda v: 3 if v>=3 else v)
ames2['garagecars'] = ames2['garagecars'].apply(lambda v: 3 if v>=3 else v)
```

```python
#finally, create a cumulative garage column
ames['garage_ord'] = round(ames[['garagetype', 'garagecars', 'garagefinish']]
                         .mean(axis=1), 2)
ames2['garage_ord'] = round(ames2[['garagetype', 'garagecars', 'garagefinish']]
                         .mean(axis=1), 2)
```

```python
ames.drop(columns=['garagetype', 'garagecars', 'garagefinish'], inplace=True)
ames2.drop(columns=['garagetype', 'garagecars', 'garagefinish'], inplace=True)
```

```python
# dummy for paved drive
ames['paveddrive'] = np.where((ames['paveddrive']=='Y'), 1, 0)
ames2['paveddrive'] = np.where((ames2['paveddrive']=='Y'), 1, 0)
```

```python
# learn about the numeric fireplace columnas it'll be combined with the categorical
ames.fireplaces.value_counts()
```

```python
# it contains values 0-4 so it can be analyzed with our cat_info function
cat_info('fireplaces')
```

```python
# translate fireplace quality codes into ordered numeric values. make sure that the scale
# and the incremental value of 1 point increase mirrors the numerical 'fireplaces' columns

# after studying the firplace quality and fireplace number columns, the ordinal numeric 
# scale is found that weighs both columns on equal terms
codes = sorted(ames['fireplacequ'].unique())
ranks = [2.5, 0.5, 1.5, 0, 0, 1]
hood_dict = dict(zip(codes, ranks))
hood_dict
```

```python
# change the fireplacequ column's values from codes to numeric
ames['fireplacequ'] = ames['fireplacequ'].map(hood_dict)
ames2['fireplacequ'] = ames2['fireplacequ'].map(hood_dict)
```

```python
ames['fireplaces'] = ames['fireplaces'].apply(lambda v: 1.5 if v>=2 else v)
ames2['fireplaces'] = ames2['fireplaces'].apply(lambda v: 1.5 if v>=2 else v)
```

```python
ames['fireplaces'].value_counts()
```

```python
# creat fireplace_ord supercolumn containing both fireplace quantitative and qualitative info
ames['fireplace_ord'] = round(ames[['fireplacequ', 'fireplaces']]
                         .mean(axis=1), 2)

ames2['fireplace_ord'] = round(ames2[['fireplacequ', 'fireplaces']]
                         .mean(axis=1), 2)
```

```python
ames.drop(columns=['fireplacequ', 'fireplaces'], inplace=True)
ames2.drop(columns=['fireplacequ', 'fireplaces'], inplace=True)
```
